```python
import pandas as pd
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'c': ['C0', 'C1', 'C2', 'C3']})
left
```

```python
right = pd.DataFrame({'A': ['A0','A2', 'A3','A4'],
                     'B': ['B0','B2', 'B3','B4'],
                     'C': ['C0','C2', 'C3','B4']})
right
```

```python
intersection = pd.merge(left,right)
intersection
```

```python
inter = pd.merge(left,right, on=['A','B','C'],how='outer')
inter
```

```python
union = pd.concat([left,right])
union
```

```python
todos = pd.concat([left,right]).drop_duplicates().reset_index(drop=True)
todos
```

```python
columnaD = pd.DataFrame({'A': ['A0','A1','A2', 'A3'],
                     'B': ['B0','B1','B2', 'B3'],
                     'C': ['C0','C1','C2', 'C3'],
                     'D': ['D0','D1','D2', 'D3']})
columnaD
columnaE = pd.DataFrame({'A': ['A0','A1','A3','A4'],
                     'B': ['B0','B1', 'B3','B4'],
                     'C': ['C0','C1', 'C3','B4'],
                     'E': ['E0','E1', 'E3','E4']})

columnaE
```

```python
todos_E = pd.merge(columnaE,todos,on=['A','B','C'],how='outer' )
todos_E
```

```python
todos_D = pd.merge(columnaD, columnaE,on=['A','B','C'],how='outer' )
todos_D
```
