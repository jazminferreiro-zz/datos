```python
port_data = port_data.rename(columns={'G1':'G1_portugues'})
port_data = port_data.rename(columns={'G2':'G2_portugues'})
port_data = port_data.rename(columns={'G3':'G3_portugues'})
port_data = port_data.rename(columns={'paid':'paid_portugues'})
port_data = port_data.rename(columns={'failures':'failures_portugues'})

math_data = math_data.rename(columns={'G1':'G1_mate'})
math_data = math_data.rename(columns={'G2':'G2_mate'})
math_data = math_data.rename(columns={'G3':'G3_mate'})
math_data = math_data.rename(columns={'paid':'paid_mate'})
math_data = math_data.rename(columns={'failures':'failures_mate'})
```

```python
#hay 100 alumnos que estan en las dos clases
intersection = pd.merge(port_data,math_data,how='inner', on=['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu',
       'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime',
       'studytime', 'schoolsup', 'famsup',
       'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel',
       'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences'])
intersection.shape

```

```python
#todos los alumnos sin repetir los cien que van a las dos clases
students = pd.merge(port_data,math_data,how='outer', on=['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu',
       'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime',
       'studytime', 'schoolsup', 'famsup',
       'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel',
       'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences'])
students.shape

```

```python
students['promedio_mate']= (students.G1_mate*5 +students.G2_mate*5 + students.G3_mate*5)/3
students['promedio_portugues']= ( students.G1_portugues*5+students.G2_portugues*5+students.G3_portugues*5 )/3
students.tail()
```

```python
import numpy as np
def promedio_total(row):
   if (np.isnan(row['promedio_portugues'])):
      return row['promedio_mate']
   else:
    if(np.isnan(row['promedio_mate'])):
           return row['promedio_portugues']
    else:
            return (row.promedio_portugues + row.promedio_mate) /2
        
students['promedio_total'] =students.apply(lambda row: promedio_total(row),axis=1)
students   
        
```
